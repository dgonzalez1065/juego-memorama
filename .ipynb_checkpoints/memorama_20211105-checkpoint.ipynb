```python
import random

def inicio():
    print("""
    Bienvenido a Memorama.
    Es un juego de memoria en donde tendr√° oportunidad 
    de seleccionar dos cartas por turno.
    Si las dos cartas coinciden permanecer√°n descubiertas. 
    Tiene un n√∫mero limitado de oportunidades por lo que 
    que no deber√° de desperdiciarlos.""")
    input('Ingrese cualquier numero para continuar: ')

def impresion_cartas(cartas, mostrar):
    print()
    for indice, carta in enumerate(cartas):
        if mostrar[indice] == False:
            print('\N{black medium square}', end=' ')
        else:
            print(carta, end=' ')
        if (indice +1)%4 == 0:
            print()

def seleccion_carta(mostrar):
    oportunidad = 0
    while True:
        oportunidad = input('Seleccione una carta: ')
        if oportunidad.isnumeric() == False:
            print('Seleccion invalida. Por favor, intente de nuevo.')
            continue
        oportunidad = int(oportunidad)
        if oportunidad < 1 or oportunidad > 16:
            print('Seleccion invalida. Por favor, intente de nuevo.')
            continue
        oportunidad -= 1
        if mostrar[oportunidad]:
            print('Seleccion invalida. Por favor, intente de nuevo.')
            continue
        return oportunidad
        
cartas = ['\N{alien monster}', '\N{alien monster}', 
           '\N{red apple}', '\N{red apple}', 
           '\N{purple heart}', '\N{purple heart}', 
           '\N{shamrock}', '\N{shamrock}', 
           '\N{monkey face}', '\N{monkey face}', 
           '\N{pig face}', '\N{pig face}', 
           '\N{chicken}', '\N{chicken}', 
           '\N{honeybee}', '\N{honeybee}']
random.shuffle(cartas)
mostrar = [False]*16
pares = 0
turnos = 0
fallas = 0

inicio()

while pares != 8:
    impresion_cartas(cartas, mostrar)
    oportunidad_1 = seleccion_carta(mostrar)
    mostrar[oportunidad_1] = True
    
    impresion_cartas(cartas, mostrar)
    oportunidad_2 = seleccion_carta(mostrar)
    mostrar[oportunidad_2] = True
    
    impresion_cartas(cartas, mostrar)
    
    if cartas[oportunidad_1] == cartas[oportunidad_2]:
        print('Coindicen. Esta bien')
        pares += 1
    else:
        print('No coinciden.')
        
        mostrar[oportunidad_1] = False
        mostrar[oportunidad_2] = False
        
    turnos += 1
    
    if pares != 8:
        pregunta = input('Ingerese 0 para Parar. Presione ENTER para Continuar')
        if pregunta == '0':
            print("""
            ¬°Gracias por jugar!""")
            break
    else:
        print("""
                     ¬°Felicidades!
                      ¬°Bien hecho!""")
        print(f'Numero de turnos: {turnos}')
```
