```python
import random
def crear_matriz(num):
    matriz = []
    a = 0
    for i in range(4):
        matriz.append([])
        for j in range(4):
            matriz[i].append(num[a])
            a += 1
    
    for i in range(len(matriz)):
        print(matriz[i])
    return matriz

def crea_tablero():
    tablero = [['-' for i in range(0, 4)] for j in range(0,4)]
    
    for i in tablero:
        for j in i:
            print(j, " ", end=" ")
        print()
    return tablero

def mostrar(a, b, tablero, matriz):
    tablero[b-1][a-1] = matriz[a-1][b-1]
    for i in tablero:
        for j in i:
            print(j, " ", end = " ")
        print()
    
    return tablero[b-1][a-1]

def limpiar(a1, b1, a2, b2):
    tablero[b1-1][a1-1] = "-"
    tablero[b2-1][a2-1] = "-"
    for i in tablero:
        for j in i:
            print(j, " ", end=" ")
        print()

def validar(a, b, a2, b2):
    if  a < 1 or a > 4 or b < 1 or b > 4 or tablero[b-1][a-1] != "-":
        print("Movimiento invalido")
        return False
    elif a == a2 and b == b2:
        print("Ya fue seleccionada esa carta")
        return False
    else:
        return True


def verificar_movimientos():
    guiones = 0
    for i in tablero:
        for j in i:
            if j == "-":
                guiones += 1
    return guiones

import time
numeros = ['\N{alien monster}', '\N{alien monster}', 
           '\N{red apple}', '\N{red apple}', 
           '\N{purple heart}', '\N{purple heart}', 
           '\N{shamrock}', '\N{shamrock}', 
           '\N{monkey face}', '\N{monkey face}', 
           '\N{pig face}', '\N{pig face}', 
           '\N{chicken}', '\N{chicken}', 
           '\N{honeybee}', '\N{honeybee}']
print("Empieza el Memorama")
#for i in range(1,19):
#    numeros.append(i)
#    numeros.append(i)
random.shuffle(numeros)
matriz = crear_matriz(numeros)
tablero = crea_tablero()
jugando = True
ganador = False
guiones = 0
seguir = False
while jugando == True and ganador == False:
    valido1 = False
    valido2 = False
    while valido1 == False:
        a1 = int(input("Carta en a del primer numero: "))
        b1 = int(input("Carta en b del primer numero: "))
        valido1 = validar(a1, b1, -1, -1)
    print("Elegiste ", matriz[a1-1][b1-1])
    while valido2 == False:
        a2 = int(input("Carta en a del segundo numero: "))
        b2 = int(input("Carta en b del segundo numero: "))
        valido2 = validar(a2, b2, a1, b1)
    print("Elegiste ", matriz[a2-1][b2-1])
    tabla1 = mostrar(a1, b1, tablero, matriz)
    print("Primer numero")
    tabla2 = mostrar(a2, b2, tablero, matriz)
    print("Segundo numero")
    if tabla1 == tabla2:
        print("Esta bien")
    else:
        print("Esta mal")
        limpiar(a1, b1, a2, b2)
    if verificar_movimientos() == 0 and ganador == False:
        seguir = True
    while seguir == False:
        resp = input("Quieres continuar con el juego [Y/N]: ")
        if resp == 'Y' or resp == 'y':
            jugando = True
            seguir = True
        elif resp == 'N' or resp == 'n':
            jugando = False
            seguir = True
        else:
            seguir = False
```
